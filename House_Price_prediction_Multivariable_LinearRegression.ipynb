```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('house.csv')
df.head()
```

```python
# Here there are missing values in the daraset, so we have t filter it
# Here there is a null value in bedrooms, we have to replace the null value with median of column bedrooms
# Convert the 'bedrooms' column to numeric, handling errors
df['bedrooms'] = pd.to_numeric(df['bedrooms'], errors='coerce')  # The errors='coerce' argument replaces non-convertible values (like empty strings) with NaN (Not a Number).
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)  # Fill missing values with the median
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
